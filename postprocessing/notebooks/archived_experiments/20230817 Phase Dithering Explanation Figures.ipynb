```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.001
ts = np.arange(0, 1, dt)
sig = np.sin(4*2*np.pi*ts)
noise = 0.1 * np.sin(10*2*np.pi*ts)

combined = noise
combined[int(0.1/dt):int(0.5/dt)] = sig[:int(0.4/dt)]

sig = np.sin(4*2*np.pi*ts + 1)
noise = 0.1 * np.sin(10*2*np.pi*ts)

combined2 = noise
combined2[int(0.1/dt):int(0.5/dt)] = sig[:int(0.4/dt)]

combined2_shifted = np.zeros_like(combined2)
combined2_shifted[int(0.04/(dt)):] = combined2[:-int(0.04/(dt))]

fig, ax = plt.subplots(figsize=(8,3))
#ax.plot(ts, combined)
#ax.plot(ts, combined2_shifted, color='orange')

ax.plot(ts, (combined + combined2_shifted) / 2, color='red')

ax.set_xlabel('Time')
ax.set_ylabel('Signal Amplitude')
#ax.set_ylim(-1,1)
```
